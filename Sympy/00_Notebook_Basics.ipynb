# De basis van Jupyter Notebooks

## Opstarten

De Jupyter notebook server is het programma dat ervoor zorgt dat je de werkbladen die horen bij het vak Calculus kan openen. Er zijn twee manieren om de notebook server te starten.
 - Als eerste kan je (op Windows) de snelkoppeling gebruiken die bij een installatie van Anaconda gemaakt wordt.
 - Op macOS of Linux kan je een terminal openen en het commando `jupyter notebook` uitvoeren. Deze methode werkt ook in Windows via de Anaconda prompt.

## Het dashboard

Zodra je de notebook server opstart, zal je browser het dashboard openen. Het dashboard is de startpagina van waaruit je je werkbladen kan openen en beheren. Op het scherm zal je de bestanden in de startmap te zien krijgen. Hieronder is een voorbeeld te zien van hoe het dashboard eruit ziet:

Als je de notebook server opent via een terminal, is de startmap _altijd_ de map waarin het commando `jupyter notebook` uitgevoerd wordt.

Om een nieuw werkblad te maken, kan je op de knop `New` rechtsboven klikken (zie afbeelding hieronder) en kan je een "kernel" kiezen. Deze kernel zegt met welke programmeertaal je werkt. Standaard staat hier enkel Python in maar je kan zelf kernels installeren om verschillende programmeertalen te ondersteunen. Je zal dus minder opties te zien krijgen als op deze afbeelding te zien.

## Overzicht van de interface

Als je een nieuw werkblad maakt of een bestaand werkblad opent, krijg je de gebruikersinterface te zien. Hier kan je zelf code schrijven en uitvoeren en kan je zelf de opmaak verzorgen van het document. De interface bestaat uit volgende gebieden:

* Menubalk
* Werkbalk
* Werkbladcellen

Het werkblad heeft een interactieve rondleiding van deze elementen. Je kan deze starten in het menu "Help:User Interface Tour".

## Verschillende modi

Het werkblad werkt in twee verschillende modi. Je hebt de Edit Mode en de Command Mode. Afhankelijk van in welke modus je je bevindt zal je andere dingen kunnen doen om het werkblad vorm te geven.

### Edit mode

Edit mode wordt gekarakteriseerd door een groene rand van een cel en een knipperende cursor in de cel zelf:

<img src="images/edit_mode.png">

Als een cel in edit mode staat, kan je er tekst in typen zoals bij een gewone tekstverwerker.

<div class="alert alert-success">
Je komt in edit mode terecht door in de linkermarge te klikken en op `Enter` te duwen. Voor een cel met code kan je ook klikken in het invoervak zelf en bij tekstcellen kan je ook dubbelklikken in de tekst.
</div>

### Command mode

Command mode wordt gekarakteriseerd door een grijze rand van de cel en een blauwe linkermarge:

<img src="images/command_mode.png">

Als je je in command mode bevindt, kan je het werkblad als geheel bewerken maar kan je niet typen in een individuele cel. Heel belangrijk is hierbij dat in command mode sommige toetsen van je toetsenbord veranderen in snelkoppelingen zodat je snel en efficiënt aanpassingen kan maken. Als je bijvoorbeeld een cel selecteert in command mode en je duwt op de toets `c`, dan kopieer je de hele cel.

## Navigatie met de muis

Alles wat je nodig hebt om te navigeren kan je via de muis vinden in de menubalk en de werkbalk. Beide zijn te vinden bovenaan:

<img src="images/menubar_toolbar.png" width="786px" />

Het basisidee van navigatie met de muis is dat **cellen geselecteerd kunnen worden door erop te klikken**. Afhankelijk van de modus waarin je je bevindt zal de linkerrand van de cel groen of blauw worden (zoals hierboven uitgelegd).

Het tweede idee achter navigatie met de muis is dat **acties die op cellen kunnen worden uitgevoerd meestal beperkt blijven tot de huidige geselecteerde cel**. Als je dus een cel met code hebt en deze code wilt uitvoeren, klik je op de knop 'Play' in de werkbalk of ga je via "Cell:Run Cells" in de menubalk. Als je dan weer een cel wilt kopiëren doe je dit via de knop 'Kopiëren' in de werkbalk of via "Edit:Copy" in de menubalk. 

Cellen die tekst en hoofdingen bevatten hebben nog een extra toestand waarin ze kunnen verkeren. Ze kunnen gerenderd of ongerenderd voorkomen. Dit wilt zeggen dat iedere cel met tekst erin eigenlijk wordt opgebouwd via een minimaal opmaaksysteem genaamd Markdown. Als je dubbelklikt op een cel met tekst, krijg je de 'broncode' van deze cel te zien. Door de cel uit te voeren via de knop in de werkbalk (of via "Cell:Run Cells" in de menubalk) wordt deze broncode omgezet in een mooi geformatteerd stuk tekst. Markdown heeft een eenvoudige syntax die je [op het internet](https://daringfireball.net/projects/markdown/syntax) in zijn geheel kunt vinden. De werkbladen voor Calculus bevatten echter de meeste elementen van Markdown die je ooit zou willen gebruiken voor je eigen opmaak. Als je dus bezig bent in de werkbladen voor dit vak en je vraagt je af hoe een bepaald stuk tekst precies geformatteerd werd, kan je altijd op de tekst dubbelklikken om dit te bekijken.

Nu we toch nog bezig zijn over de cellen met tekst vermelden we er best nog even bij dat deze werkbladen ook ondersteuning hebben voor LaTeX. Met enkele dollartekens kan je een wiskundige uitdrukking middenin een zin zetten. Als je mensen er dus aan wilt herinneren dat $e^{i\pi}+1=0$ dan maken deze werkbladen daar geen probleem van. Voor geïsoleerde vergelijkingen kan je dubbele dollartekens gebruiken. Zo kan je uitgebreidere uitdrukkingen meegeven zoals

$$y = \frac{\sin\left(\theta^2\right)}{\theta^3}.$$

Dubbelklik eens op de bovenstaande vergelijking om naar edit mode te gaan. Hier zie je de LaTeX-syntax die gebruikt is. Je kan de cel terug uitvoeren door op `run` bovenaan te klikken.

## Navigatie via het toetsenbord

De reden waarom we met verschillende modi werken in werkbladen is zodat het toetsenbord efficiënt gebruikt kan worden. We hebben bijgevolg twee sets toetsencombinaties: één set voor edit mode en één set voor command mode.

De belangrijkste toetsen zijn `Enter`, waarmee je van command mode naar edit mode gaat, en `Esc`, waarmee je van edit mode naar command mode gaat.

In edit mode wordt het grootste deel van het toetsenbord voorbehouden om te kunnen typen in een cel. Er zijn dus relatief weinig speciale toetsencombinaties voor edit mode. In command mode is het hele toetsenbord beschikbaar om operaties mee uit te voeren. Het menu `Help`->`Keyboard Shortcuts` toont een lijst met beschikbare combinaties.

Volgende toetsen in command mode kunnen er al voor zorgen dat je je muis bijna niet nodig hebt

1. Basis navigatie: `enter` gaat naar edit mode, `ctrl-enter` voert een cel uit, `alt-enter` voert een cel uit en plaatst er een lege codecel onder, met `pijl omhoog` en `pijl omlaag` navigeer je tussen de verschillende cellen.
2. Werkblad opslaan: `s`
2. Type van cel veranderen: `y` maakt van een cel een codecel, `m` maakt van een cel een tekstcel.
3. Cel aanmaken: `a` plaatst een lege codecel *boven* de huidige cel, `b` plaatst een lege codecel *onder* de huidige cel
4. Cell bewerken: `x` knipt de geselecteerde cellen, `c` kopiëert de geselecteerde cellen, `v` plakt de gekopiëerde cellen onder de huidige cel, `d` verwijdert de geselecteerde cellen en `z` maakt een verwijdering ongedaan.
5. Kernel operaties: tweemaal op `0` (op een QWERTY-klavier) of tweemaal op `à` (op een AZERTY-klavier) drukken herstart de kernel. Hierdoor wordt het geheugen gewist en kan je met een schone lei herbeginnen aan je code.

## Iets meer over de kernel

Iedere keer dat je een notebook opent, wordt er een zogenaamde kernel mee opgestart. Je ziet rechtsboven in je scherm welke kernel er nu actief is. In de pc-sessies moet dit steeds `python3` zijn. Als dit niet het geval is, ga je via `Kernel: Change Kernel: Python[python3]`. Op je eigen pc zal je niets speciaals moeten doen (als je tenminste Anaconda voor Python 3.7 hebt geïnstalleerd).

De kernel zorgt ervoor dat al je Python-code wordt uitgevoerd. De kernel houdt ook al je variabelen bij. Moest het voorvallen dat je variabelen een soep worden en je graag met een schone lei herbegint, kan je de kernel herstarten door het commando uit de lijst hierboven of via `Kernel: Restart`. Je kan ook `Kernel: Restart & Clear Output` of `Kernel: Restart & Run All` gebruiken om alle code-output te verwijderen of meteen alle cellen uit te voeren. Alle code die je al hebt getypt, wordt behouden. Het is enkel het interne geheugen dat wordt gewist.

## Opslaan en afdrukken

Een notebook wordt standaard automatisch opgeslagen. Als je tijdens de oefenzittingen een kopie van het lege werkblad wilt behouden, zijn er twee manieren:
 * Ofwel open je het werkblad en ga je naar `File:Make a Copy`. Er opent dan een nieuw werkblad dat dezelfde naam heeft als het oorspronkelijke maar met `-Copy1` erachter. Je kan dan hier je aanpassingen maken zonder het oorspronkelijke werkblad te veranderen.
 * Ofwel open je het werkblad en ga je naar `File:Save and Checkpoint`. Je ziet dan tussen je bestanden een map met de naam `.ipynb_checkpoints` verschijnen waarin al je checkpoints worden opgeslagen. Je kan dan terugkeren naar een vorige versie via `File:Revert to Checkpoint`.
 
Let wel op: je kan maar één checkpoint gebruiken per werkblad. Het is dus aangeraden op via de eerste methode te werken.

Hoe kan je een notebook opslaan onder andere formaten? Via `File:Download as` kan je een notebook in verschillende formaten downloaden. De handigste hiervan zijn
* `HTML (.html)`: Hiermee kan je de volledige notebook in je browser bekijken zonder dat je de code of de tekstvakken kan aanpassen. Dit is ideaal als je de resultaten wilt delen met iemand die niet alle nodige software ter beschikking heeft.
* `Python (.py)`: Hiermee download je alle code als een Python-script. De begeleidende tekst komt dan gewoon in comments te staan.

De optie pdf is niet ondersteund in de PC-klassen.
