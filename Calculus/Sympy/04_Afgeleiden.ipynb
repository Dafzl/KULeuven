**Opdracht 1:** Begin met Sympy in te laden, enkele standaardinstellingen zoals het gebruik van pretty printing in te schakelen, en enkele symbolen (zoals $x$, $y$, $z$ en $t$) vast te leggen.

```python
import sympy as sp
x, y, z, t = sp.symbols('x y z t')
```

## Afleiden van uitdrukkingen

Sympy heeft twee commando's om afgeleiden te berekenen van uitdrukkingen. Het *actieve* commando `diff(uitdrukking, variabele)` leidt de uitdrukking af, terwijl het *inerte* commando `Derivative(uitdrukking, variabele)` commando de te berekenen afgeleide teruggeeft in de standaardnotatie.

Voer onderstaande blokken code uit om de verschillende commando's aan het werk te zien.

```python
afg = sp.Derivative(sp.sin(1/sp.sqrt(x)), x)
afg
```

```python
afg.doit()
```

```python
sp.diff(sp.sin(1/sp.sqrt(x)), x)
```

**Opdracht 2:** Leid de functie met voorschrift $f(y) = \frac{e^{2y}}{y^2-7}$ af naar $y$.

```python
f = sp.Function('f')
f = sp.exp(2*y)/(y**2 - 7)
sp.diff(f, y)
```

**Opdracht 3:** De uitdrukking $x^t$ kunnen we zien als een machtsfunctie in de variabele $x$, of als een exponentiële functie in de variabele $t$. Bereken achtereenvolgens de afgeleide van deze uitdrukking naar $x$ en naar $t$.

```python
f = x**t
```

```python
sp.diff(f, x)
```

```python
sp.diff(f, t)
```

Hogere-orde afgeleiden kunnen ook met Sympy worden berekend. Hiervoor bestaan twee mogelijke commando's die hetzelfde resultaat opleveren.

**Opdracht 4:** Voer onderstaande commando's uit.

```python
trigexpr = x*sp.sin(x)  # Hogere-orde afgeleide
sp.diff(trigexpr, x, 2)
```

```python
sp.diff(trigexpr, x, x)
```

**Opdracht 5:** Bereken de eerste, tweede, en 2018e afgeleide van de functie met voorschrift $f(x) = x e^x$. Kun je op basis hiervan een vermoeden formuleren wat de $n$-de afgeleide van de functie zal zijn?

```python
f = x * sp.exp(x)
sp.diff(f, x, 2018)
```

**Opdracht 6:** We beschouwen opnieuw de stuksgewijs gedefinieerde functie
$$ h: \mathbb{R} \rightarrow \mathbb{R}: x \mapsto
\begin{cases}
x+2 \quad \quad x \leq 1 \\
x^2 \qquad \quad x > 1.
\end{cases} $$
Definieer deze functie in Sympy, en bereken de afgeleide functie $h1$ gebruik makende van wat je zonet geleerd hebt. Test of je code het juiste resultaat geeft door $h1(x)$ en $h1(3)$ te laten printen.

```python
h = sp.Function('h')
h = sp.Piecewise((x + 2, x <= 1), (x**2, x > 1))
h
```

```python
h1 = sp.Function('h1')
h1 = sp.diff(h, x)
h1
```

```python
h1.subs(x, 3)
```

## Rekenregels voor afgeleiden

Ook als er geen concreet functievoorschrift voorhanden is, kan Sympy nog steeds de rekenregels voor afgeleiden toepassen om uitdrukkingen met afgeleiden zo eenvoudig mogelijk te schrijven.

```python
u = sp.Function('u')
v = sp.Function('v')
```

**Opdracht 7:** We zullen nagaan of Sympy de voornaamste rekenregels in verband met afgeleiden kan toepassen.
- Bereken de afgeleide van $u+v$. Kent Sympy de somregel?
- Bereken de afgeleide van $uv$. Kent Sympy de productregel?
- Bereken de afgeleide van $u/v$. Kent Sympy de quotiëntregel?
- Bereken de afgeleide van $u \circ v$. Kent Sympy de kettingregel?

```python
sp.diff(u(x)+v(x), x)
```

```python
sp.diff(u(x)*v(x), x)
```

```python
sp.diff(u(x)/v(x), x)
```

```python
sp.diff(u(v(x)), x)
```
