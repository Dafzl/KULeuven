## Vectoren

Als je een kolomvector wilt maken, heeft iedere rij maar één element. Als je één lijst meegeeft aan `Matrix` interpreteert Sympy dit als een kolomvector.

```python
v = sp.Matrix([2, 1, 0.5])
display(v)
```

Een rijvector bevat maar één rij en kan dus op volgende manier gemaakt worden

```python
u = sp.Matrix([[2, 1, 0.5]])
display(u)
```

Een tweede manier om een rijvector aan te maken is door eerst een kolomvector te maken en deze te transponeren. Om een matrix (en dus in het bijzonder een kolomvector) te transponeren kan je volgende code gebruiken.

```python
u_kolom = sp.Matrix([2, 1, 0.5])
display(u_kolom)
display(u_kolom.T)
```

Tenslotte weet je uit lineaire algebra dat je de norm van de vector `w` zou kunnen berekenen via de code

``sp.sqrt(w.dot(w))``

maar volgende manier is iets korter en overzichtelijker.

```python
w.norm()
```

Een waarschuwing is nog op zijn plaats. De functie `norm()` veronderstelt dat de elementen van de vectoren complexe getallen zijn. Omdat de norm van vectoren in $\mathbb{C}^n$ algemener gedefinieerd is dan de norm van vectoren in $\mathbb{R}^n$, kunnen soms onverwachte resultaten ontstaan.

```python
w4 = sp.Matrix([1/sp.sqrt(1-sp.cos(t)), 1])

display(w4)

# Norm berekend in de veronderstelling dat de vector complexe elementen bevat.
display(w4.norm())   # Sympy werkt het kwadraat van de modulus van de eerste term niet verder uit.

# Norm berekend via scalair product
sp.sqrt(w4.dot(w4))
```

De cosinus en sinus van een complex getal liggen niet per se tussen -1 en 1. Daardoor kan er ambiguïteit zijn in het evalueren van de modulus van dit getal en laat Sympy dit onuitgewerkt staan.

Anderszijds geeft de implementatie via het scalair product niet het juiste resultaat als er met complexe getallen gerekend moet worden.

Wees dus kritisch in het interpreteren van computergegenereerde berekeningen. In de context van normen is het meestal een goed idee om via het scalair product te werken, als je enkel reële getallen beschouwt, en anders `.norm()` te gebruiken.

## Meer over matrices

Voor matrices werken de operatoren `+`, `*` en `**` zoals verwacht. Je kan hiermee matrices optellen, vermenigvuldigen of tot een macht verheffen. Om de inverse van een matrix te berekenen, verhef je deze gewoon tot de macht -1.

```python
M = sp.Matrix([[1, 3], [-2, 3]])
N = sp.Matrix([[0, 3], [0, 7]])
```

Optelling:

```python
M + N
```

Vermenigvuldiging met een getal:

```python
5 * M
```

Vermenigvuldiging met een matrix:

```python
M * N
```

Machtsverheffing:

```python
M ** 2
```

Inverse:

```python
M ** (-1)
```

Tot slot kan Sympy ook de eigenwaarden en eigenvectoren van een matrix berekenen. Dit kan je doen door op een matrix de methoden `.eigenvals()` en `.eigenvects()` aan te roepen.

```python
M = sp.Matrix([[3, -2,  4, -2], [5,  3, -3, -2], [5, -2,  2, -2], [5, -2, -3,  3]])
M.eigenvals()
```

Zoals je kan zien, is de output van de methode `.eigenvals()` een dictionary. Bij iedere eigenwaarde staat de bijhorende multipliciteit gegeven. Om de multipliciteiten weg te laten, kan je van deze dictionary een lijst maken door `list(M.eigenvals())` uit te voeren.

Om de eigenvectoren van `M` te berekenen, gebruik je

```python
M.eigenvects()
```

De uitvoer is een lijst van lijsten. Ieder element bevat een eigenwaarde, de multipliciteit van die eigenwaarde en een basis van de bijhorende eigenruimte.
